```python
# Drop unwanted columns (ad_id is not needed)
ad_df.drop(columns=['Ad_ID'], inplace=True)
ad_df.columns
```

```python
# Fix Typos in Campaign Name
ad_df["Campaign_Name"] = ad_df["Campaign_Name"].str.replace(
    r"(DataAnalyticsCourse|Data Anlytics Corse|Data Analytcis Course|Data Analytics Corse)",
    "Data Analytics Course",
    regex=True
)
ad_df['Campaign_Name'].str.capitalize()
ad_df['Campaign_Name'].unique()
```

```python
# Fix Typos in Keyword column and Capitialize each word
ad_df['Keyword'] = ad_df['Keyword'].str.replace(r"(analitics|anaytics)","Analytics",regex=True)
ad_df['Keyword'] = ad_df['Keyword'].str.title()
ad_df['Keyword'].unique()
```

```python
# Fix Inconsistency in Location Column - Capitalize and fixing Typos
ad_df['Location'] = ad_df['Location'].str.capitalize()
ad_df['Location'] = ad_df['Location'].str.replace(r"(Hyderbad|Hydrebad)","Hyderabad",regex=True)
ad_df['Location'].unique()
```

```python
# Fix  Inconsistency in Device Column - Capitalize
ad_df['Device'] = ad_df['Device'].str.capitalize()
ad_df['Device'].unique()
```

```python
# Convert data types - REmoving '$' amd convert into float data type
ad_df['Cost'] = ad_df['Cost'].str.replace(r'\$', '', regex=True).astype(float)
ad_df['Sale_Amount'] = ad_df['Sale_Amount'].str.replace(r'\$', '', regex=True).astype(float)
ad_df[['Cost','Sale_Amount']].head()
```

```python
# Fixing Data Format
from dateutil import parser

# Clean and standardize text(removing white spaces)
ad_df['Ad_Date'] = ad_df['Ad_Date'].astype(str).str.strip()

# Define a safe_parse function
def safe_parse(date_str):
    try:
        # Try parsing each date, interpret the string as a date
        return parser.parse(date_str, dayfirst=True).date()
    except Exception:
        return pd.NaT
# Apply parsing function
ad_df['Ad_Date'] = ad_df['Ad_Date'].apply(safe_parse)

ad_df['Ad_Date']
```

```python
# Decide whether to use mean or median for imputation based on skewness and outlier presence
# used function
def choose_imputation(series):
    # remove missing values
    data = series.dropna()
    # Calculate skewness to measures symmetry of the distribution
    skew = data.skew()
    # Detect outliers using IQR
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data < lower) | (data > upper)]
    # Deciding Mean or mediean
    if abs(skew) < 0.5 and len(outliers) == 0:
        method = "mean"
        value = data.mean()
    else:
        method = "median"
        value = data.median()

    return method, value

# Numeric columns except Conversion Rate(as it need seprate calculation)
numeric_cols = ['Clicks','Impressions','Cost','Leads','Conversions','Sale_Amount']

for col in numeric_cols:
    method, value = choose_imputation(ad_df[col])
    #print(f"{col}: using {method}, value = {value}")
    ad_df[col] = round(ad_df[col].fillna(value),2)
#Check for null
ad_df[['Clicks','Impressions','Cost','Leads','Conversions','Sale_Amount']].isnull().sum()
```

```python
# Recalculate Conversion Rate instead of imputing
ad_df['Conversion Rate'] = round(ad_df['Conversions'] / ad_df['Clicks'],2)
ad_df['Conversion Rate']
```

```python
# Checked isnull to ensure is Data cleaned
ad_df.isnull().sum()
```

```python
# Feature transformations (date parts, derived fields if required for analysis)
# Extract date parts which are possible
ad_df['Ad_Date'] = pd.to_datetime(ad_df['Ad_Date'], errors='coerce')
ad_df['Year'] = ad_df['Ad_Date'].dt.year
ad_df['Month'] = ad_df['Ad_Date'].dt.month
ad_df['Day'] = ad_df['Ad_Date'].dt.day
ad_df['Day_Of_Week'] = ad_df['Ad_Date'].dt.day_name()
ad_df['Week_Of_Year'] = ad_df['Ad_Date'].dt.isocalendar().week
ad_df[['Year','Month','Day','Day_Of_Week','Week_Of_Year']]
```

```python
# Derived Metrics
# Calculate Cost per Acquisition (CPA)
ad_df['CPA'] = round(ad_df['Cost'] / ad_df['Conversions'],2)

# Calculate Cost per click (CPC)
ad_df['CPC'] = round(ad_df['Cost'] / ad_df['Clicks'],2)

# Calculate Cost per lead (CPL)
ad_df['CPL'] = round(ad_df['Cost'] / ad_df['Leads'],2)

# Calculate Revenue per conversion (RPC)
ad_df['RPC'] = round(ad_df['Sale_Amount'] / ad_df['Conversions'],2)

# Calculate Sales Efficiency(Return on Spend,Revenue Efficiency,Revenue-to-Cost Ratio)
ad_df['Sales_ROI'] = round(ad_df['Sale_Amount'] / ad_df['Cost'],2)

# Calculate CTR (Click-Through Rate)
ad_df['CTR'] = round(ad_df['Clicks'] / ad_df['Impressions'],2)

# Calculate CTR (%)
ad_df['CTR%'] = round((ad_df['Clicks'] / ad_df['Impressions']) * 100,2)

# Calculate Conversion Rate (CVR)
ad_df['CVR'] = round(ad_df['Conversions'] / ad_df['Clicks'],2)
# CVR (%)
ad_df['CVR%'] = round((ad_df['Conversions'] / ad_df['Clicks']) * 100,2)

# Calculate Lead Conversion Rate (LCR)
ad_df['LCR'] = round(ad_df['Conversions'] / ad_df['Leads'],2)
# Calculate Lead Conversion Rate Percent (LCR%)
ad_df['LCR%'] = round(ad_df['LCR'] * 100,2)

# Calculate Profit (Return on Investment)
ad_df['Profit'] = round((ad_df['Sale_Amount'] - ad_df['Cost']) / ad_df['Cost'],2)
# Calculate Profit%
ad_df['Profit%'] = round(((ad_df['Sale_Amount'] - ad_df['Cost']) / ad_df['Cost']) * 100,2)

ad_df[['CPA','CPC','CPL','RPC','Sales_ROI','CTR','CTR%','CVR','CVR%','LCR','LCR%','Profit','Profit%']]
```
