# Project Title

Clicks to Conversions: Google Ads Sales Analysis through Python-Powered Insights and Power BI Visualization

# **Objective(Problem statment)**

1. To uncover meaningful patterns, trends, and insights from Google Ads campaign data through exploratory data analysis (EDA).  

2. To clean, transform, and preprocess the dataset for accurate, reliable, and structured analysis.  

3. To design clear and impactful visualizations in Power BI that effectively communicate ad performance and sales outcomes.  

4. To evaluate key performance indicators (KPIs) such as CTR (Click Through Rate), CPC (Cost per Click), ROAS (Return on Ad Spend), and overall campaign ROI.  

5. To compare campaign performance across dimensions such as Location, Device, and Keyword, identifying the most effective targeting strategies.  

6. To document the workflow and highlight limitations of the dataset and analysis, ensuring reproducibility, transparency, and credibility of findings.  


# Outcome

The project uncovers key patterns, trends, and hidden insights from Google Ads campaign data.Through systematic cleaning, preprocessing, and exploratory analysis, the dataset provides a clear understanding of ad performance across metrics such as clicks, impressions, conversions, and sales revenue. Interactive Power BI dashboards highlight campaign effectiveness by location, device, and keyword, enabling evidence based decision making. #**Dataset Information**

**Source:**  [Google Ads Sales Analysis](https://github.com/narmada-j/Main-Project/blob/main/GoogleAds_DataAnalytics_Sales_RawData.csv)

**Year / Timeline:**  2024

**Dataset Column/features Description:**

* **Ad_ID** - A unique identifier for each ad campaign.  

* **Campaign_Name** - The name of the advertising campaign.

* **Clicks** - The number of clicks the ad received.  

* **Impressions** - The total number of times the ad was displayed.

* **Cost** - The total expenditure on the ad campaign.  

* **Leads** - The number of potential customers generated.  

* **Conversions** - The number of actual conversions (such as signups or sales).  

* **Conversion Rate** - The percentage of clicks that resulted in conversions.

* **Sale_Amount** - The revenue generated from conversions.  

* **Ad_Date** - The date on which the ad activity occurred.  

* **Location** - The city where the ad was shown.  

* **Device** - The type of device used (Mobile, Desktop, or Tablet).  

* **Keyword** - The search keyword that triggered the ad.  


# **Note For Kind of Analysis to be done**

**Type of Analysis**

* Descriptive Analysis (summarizing the dataset)
  * Summarizes the Google Ads campaign dataset to provide an overall view of performance

  * Analyzes key metrics such as clicks, impressions, conversions, cost, and sales revenue using summary statistics (mean, median, maximum, minimum)

  * Applies frequency counts to measure occurrences, for example the number of campaigns by Location or Device

  * Uses simple visualizations such as bar charts, pie charts, and line graphs to present campaign performance and highlight trends

  * Outcome: Provides a comprehensive snapshot of “what happened” across campaigns

* Diagnostic Analysis (finding reasons behind patterns)
  *  Investigates the underlying reasons behind observed patterns in campaign performance

  * Uses correlation analysis to explore relationships between variables, such as the link between advertising cost and conversions

  * Applies comparative analysis to evaluate performance across dimensions like Location, Device, and Keyword

  * Employs drill down dashboards in Power BI to isolate underperforming campaigns and uncover specific areas for improvement

  * Outcome: Provides actionable insights into “why it happened”, guiding optimization strategies such as reallocating budget to high performing keywords or devices
